## Mathematical theory

Ancient Egyptians used fractions in a different way than I do today.
Instead of writing fractions like 3/4 or 5/6, they wrote them as sums
of unit fractions.

A unit fraction is a fraction of the form:

1/n

where n is a positive integer.

For example:

2/3 = 1/2 + 1/6

3/4 = 1/2 + 1/4

5/6 = 1/2 + 1/3

One important result in mathematics states that every positive rational
number can be written as a sum of unit fractions.

One common way to find such a representation is the greedy algorithm.

## Greedy Algorithm

One common method for representing fractions as Egyptian fractions is
the greedy algorithm.

The idea behind the greedy algorithm is simple. At each step is selected
the largest possible unit fraction that is smaller than or equal to
the remaining fraction.

For example, if it's needed to represent the fraction 4/13, the algorithm
first selects the unit fraction 1/4 because it is the largest unit
fraction that does not exceed 4/13.

Then it's subtracted 1/4 from 4/13 and repeat the process with the
remaining fraction.

This process continues until the remaining fraction becomes zero.

## Python Implementation

To implement the greedy algorithm I use Python together with
the built-in fractions module.

The algorithm works as follows:

1. Start with a fraction a/b
2. Compute the smallest integer n such that 1/n ≤ a/b
3. Add the unit fraction 1/n to the result
4. Subtract 1/n from the current fraction
5. Repeat until the fraction becomes zero

This approach guarantees that every positive rational number
can be written as a sum of unit fractions.

```python
from fractions import Fraction
import math

def egyptian_fraction(a, b):
    result = []
    frac = Fraction(a, b)
    while frac.numerator != 0:
        n = math.ceil(frac.denominator / frac.numerator)
        result.append(n)
        frac -= Fraction(1, n)
    return result
```
