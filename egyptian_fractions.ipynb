Egyptian fractions are fractions written as a sum of unit fractions.

2/3 = 1/2 + 1/6

In this project I implement the greedy algorithm in Python

## Mathematical theory

Ancient Egyptians used fractions in a different way than I do today.
Instead of writing fractions like 3/4 or 5/6, they wrote them as sums
of unit fractions.

A unit fraction is a fraction of the form:

1/n

where n is a positive integer.

For example:

2/3 = 1/2 + 1/6

3/4 = 1/2 + 1/4

5/6 = 1/2 + 1/3

One important result in mathematics states that every positive rational
number can be written as a sum of unit fractions.

One common way to find such a representation is the greedy algorithm.
